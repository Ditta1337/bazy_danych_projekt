```python
CONNECTION_STRING = "DRIVER={SQL Server};SERVER=your_server;DATABASE=your_database;UID=your_username;PWD=your_password"
```

### 2. Establish connection with database

```python
connection = pyodbc.connect(CONNECTION_STRING)
cursor = connection.cursor()

fake = Faker()
```

### 3. Define generating functions

```python
# example
def fill_payments(num_records):
    for i in range(num_records):
        cursor.execute("INSERT INTO payments (payment_status, payment_date, payment_amount) VALUES (?, ?, ?)",
                          fake.random_element(elements=('paid', 'unpaid')),
                          fake.date_time_between(start_date='-1y', end_date='now'),
                          fake.pydecimal(left_digits=2, right_digits=2, positive=True, min_value=0, max_value=10000))
        connection.commit()
```

### 4. Invoke functions and close connection

```python
# invoke methods
fill_payments()

# close connection
connection.commit()
connection.close()
```
